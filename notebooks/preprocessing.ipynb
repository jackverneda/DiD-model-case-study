```python
import pandas as pd
```

```python
df = pd.read_csv('../data/IHME-GBD_2021_DATA-9d1094e9-1.csv')
df.head()
```

```python
df['cause_name'] = df['cause_name'].astype(str)
df.dtypes
```

```python
df_death = df[df['measure_name']=='Deaths']
```

```python
metric_number = df_death[df_death['metric_name'] == 'Number'][['cause_name','year','val']]
metric_rate   = df_death[df_death['metric_name'] == 'Rate'][['cause_name','year','val']]
metric_percent = df_death[df_death['metric_name'] == 'Percent'][['cause_name','year','val']]
```

```python
print(metric_number.shape)
print(metric_rate.shape)
print(metric_percent.shape)
```

```python
metric_percent
```

```python
cleansed =  metric_number.merge(
    metric_rate,
    on = ['cause_name', 'year'], 
    how ='inner',
    suffixes=('_number', '_rate'), 
    validate='1:1'
).merge(
    metric_percent,
    on = ['cause_name', 'year'], 
    how ='inner',
    suffixes=('', '_percent'), 
    validate='1:1'
)
cleansed.head()
```
